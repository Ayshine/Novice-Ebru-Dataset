###  Exploratory Data Analysis of Novice Ebru Dataset

In this notebook we are trying to explore the data and plot some prior results.

```python
# Import libraries necessary for this project
import pandas as pd
import matplotlib.pyplot as plt
# Pretty display for notebooks
%matplotlib inline
```

```python
df = pd.read_csv('results.csv')
```

```python
from IPython.display import display
display(df.head())
```

```python
alpha_beta_1 = df[ (df['Content_Weight']==1) & (df['Style_Weight']==1000.0)]
alpha_beta_2 = df[ (df['Content_Weight']==8) & (df['Style_Weight']==10000.0)]
alpha_beta_3 = df[ (df['Content_Weight']==5) & (df['Style_Weight']==1000.0)]
alpha_beta_4 = df[ (df['Content_Weight']==5) & (df['Style_Weight']==10000.0)]
```

```python
alpha_beta_1["Content_Loss"].plot(label = 'α/β 1/$10^3$')
alpha_beta_2["Content_Loss"].plot(label = 'α/β 8/$10^4$')
alpha_beta_3["Content_Loss"].plot(label = 'α/β 5/$10^3$')
alpha_beta_4["Content_Loss"].plot(label = 'α/β 5/$10^4$')
plt.title("Content Loss α/β ")
plt.xlabel('Target Image')
plt.ylabel('Content Loss')
plt.legend()
plt.show()
```

```python
alpha_beta_1["Style_Loss"].plot( label = 'α/β 1/$10^3$')
alpha_beta_2["Style_Loss"].plot(label = 'α/β 8/$10^4$')
alpha_beta_3["Style_Loss"].plot(label = 'α/β 5/$10^3$')
alpha_beta_4["Style_Loss"].plot(label = 'α/β 5/$10^4$')
plt.title("Style loss  α/β")
plt.xlabel('Target Image')
plt.ylabel('Style Loss')
plt.legend()
plt.show()
```

```python
alpha_beta_1["Total_Loss"].plot(label = 'α/β 1/$10^3$')
alpha_beta_2["Total_Loss"].plot(label = 'α/β 8/$10^4$')
alpha_beta_3["Total_Loss"].plot(label = 'α/β 5/$10^3$')
alpha_beta_4["Total_Loss"].plot(label = 'α/β 5/$10^4$')
plt.title("Total Loss  α/β")
plt.xlabel('Target Image')
plt.ylabel('Total Loss')
plt.legend()
plt.show()
```
